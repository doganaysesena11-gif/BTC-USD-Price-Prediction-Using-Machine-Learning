```python
from prophet import Prophet
import yfinance as yf
```

```python
df=yf.download("BTC-USD","2000-01-01","2026-10-02")
```

```python
df.head()
```

```python
df=df[["Close"]]
df.head()
```

```python
df=df.reset_index()
```

```python
df.columns=["ds","y"]
```

```python
model=Prophet()
model.fit(df)
```

```python
future=model.make_future_dataframe(360)
```

```python
predict=model.predict(future)
```

```python
model.plot(predict)
```

```python
model.plot_components(predict)
```
